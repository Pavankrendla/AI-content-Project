```python
# importing libraries
import cv2
import os
import sys
import subprocess
import speech_recognition as sr
import youtube_dl
import datetime
import pyttsx3
from moviepy.editor import VideoFileClip
from youtubesearchpython import VideosSearch
from bs4 import BeautifulSoup
import requests, json, lxml

# converting video to audio format
def convert_video_to_audio_moviepy(video_file, output_ext="mp3"):
    """Converts video to audio using MoviePy library
    that uses `ffmpeg` under the hood"""
    filename, ext = os.path.splitext(video_file)
    clip = VideoFileClip(video_file)
    clip.audio.write_audiofile(f"{filename}.{output_ext}")
```

```python
# Initialize recognizer
r = sr.Recognizer() 

# input audio file
harvard = sr.AudioFile('harvard.wav')
with harvard as source:
    audio = r.record(source)
text = r.recognize_google(audio)
print(text)
```

```python
# Read video file
convert_video_to_audio_moviepy('english_conv.mp4', 'wav')

# Load audio file
english_conv = sr.AudioFile('english_conv.wav')

# Open audio file and perform speech recognition
with english_conv as source:
    audio = r.record(source)

# Perform speech recognition on audio
text = r.recognize_google(audio)
```

```python
# Output recognized text
print(text)
```

```python
# https://docs.python-requests.org/en/master/user/quickstart/#passing-parameters-in-urls

params = {
    "q": text, # query example
    "hl": "en",          # language
    "gl": "in",          # country of the search, UK -> United Kingdom
    "start": 0,          # number page by default up to 0
    #"num": 100          # parameter defines the maximum number of results to return.
}

# https://docs.python-requests.org/en/master/user/quickstart/#custom-headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"
}

page_limit = 10           # page limit, if you do not need to parse all pages
page_num = 0

data = []

while True:
    page_num += 1
        
    html = requests.get("https://www.google.com/search", params=params, headers=headers, timeout=30)
    soup = BeautifulSoup(html.text, 'lxml')
    
    for result in soup.select(".tF2Cxc"):
        title = result.select_one(".DKV0Md").text
        try:
           snippet = result.select_one(".lEBKkf span").text
        except:
           snippet = None
        links = result.select_one(".yuRUbf a")["href"]
      
        data.append({
          "title": title,
          "snippet": snippet,
          "links": links
        })

    if page_num == page_limit:
        break
    if soup.select_one(".d6cvqb a[id=pnnext]"):
        params["start"] += 10
    else:
        break
```

```python
google_search_responses = json.dumps(data, indent=2, ensure_ascii=False)
print(google_search_responses)
```
